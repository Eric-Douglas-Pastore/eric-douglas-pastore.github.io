## What are the most common fan engagement goals mentioned by venues?

```python
pd.options.display.max_rows=None
```

```python
golf_temp['main_fan_engagement_targets'] = golf_temp['main_fan_engagement_targets'].str.replace(' ','')
# One-hot encoding with comma as a separator
dummies = golf_temp['main_fan_engagement_targets'].str.get_dummies(sep=',')
# Concatenate the encoded columns with the original DataFrame
golf_temp = pd.concat([golf_temp, dummies], axis=1)
golf_temp=golf_temp.drop(columns={'unsure'})
```

```python
golf_fans = dummies.iloc[:, 0:len(dummies)-1]
golf_fans = pd.concat([golf_fans, golf_temp.course_name], axis=1)
golf_fans = pd.concat([golf_fans,golf_temp.main_fan_engagement_targets],axis=1)
```

```python
golf_fans = golf_fans.drop(columns={'unsure'})
golf_fans
```

```python
golf_fans = golf_fans[golf_fans['main_fan_engagement_targets'] != 'unsure']
```

```python
golf_fans = golf_fans.drop_duplicates()
```

```python
golf_fans
```

```python
value_counts1 = []
for i in range(len(golf_fans.columns)-2):
    value_counts1.append(golf_fans.iloc[:,i].value_counts())
    print(golf_fans.iloc[:,i].value_counts())
```

```python
charts = []
for i in range(len(value_counts1)):
    value_counts = value_counts1[i].reset_index()
    value_counts.columns = ['category', 'count']
    charts.append(alt.Chart(value_counts,title='Histogram of Fan Engagement Techniques').mark_bar().encode(
        x=alt.X('category:N', title='Does This venue use this fan engagement technique?'),
        y=alt.Y('count:Q', title='Count of Records'),
    ))
```
